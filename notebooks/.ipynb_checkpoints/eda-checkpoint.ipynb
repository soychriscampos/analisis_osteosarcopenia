## Objetivo 1. Identificar a los pacientes con osteosarcopenia.

Este objetivo tiene como propósito identificar dentro de la muestra a aquellos pacientes que presentan osteosarcopenia, condición que combina la presencia de sarcopenia con alteraciones en la densidad mineral ósea (osteopenia u osteoporosis).

**Variables utilizadas:**  
Se creó una nueva variable `osteosarcopenia` a partir de la combinación de las variales `sarcopenia` y `clasificacion_de_densidad_mineral`. Se consideró que un paciente presenta osteosarcopenia si cumple simultáneamente con sarcopenia y un diagnóstico óseo distinto de "normal".

Esta variable será utilizada de forma transversal en varios objetivos posteriores.

```python
cond_sarcopenia = df['nivel_de_sarcopenia'] != 'sin sarcopenia'
cond_osteopenia = df['clasificacion_de_densidad_mineral'].isin(['osteopenia', 'osteoporosis'])

df['osteosarcopenia'] = cond_sarcopenia & cond_osteopenia

df['osteosarcopenia'].value_counts()
```

## Objetivo 2. Clasificar a los pacientes con osteosarcopenia, osteoporisis u osteopenia.

Este objetivo busca clasificar a los pacientes en función de dos condiciones clínicas: sarcopenia y estado óseo. 

**Variables Utilizadas:**
- `sarcopenia`: indica la presencia o ausencia de esta condición.
- `clasificacion_de_densidad_mineral`: distingue entre osteoporosis, osteopenia y normalidad ósea.

**Criterio de análisis:**  
Se realiza un análisis de frecuencia cruzada para observar cómo se distribuyen los pacientes en estas categorías y detectar si exsiste alguna relación visible entre ellas. Esta clasificación servirá como base para posteriores comparaciones funcionales y de riesgo.

```python
# Clasificación combinada según sarcopenia y densidad mineral ósea
df['tiene_osteopenia'] = df['clasificacion_de_densidad_mineral'] == 'osteopenia'
df['tiene_osteoporosis'] = df['clasificacion_de_densidad_mineral'] == 'osteoporosis'
df['tiene_sarcopenia'] = df['nivel_de_sarcopenia'] != 'sin sarcopenia'
df['osteosarcopenia'] = df['tiene_sarcopenia'] & (df['tiene_osteopenia'] | df['tiene_osteoporosis'])
```

```python
# conteo de pacientes por condicion
conteo_condiciones = pd.Series({
    'Osteosarcopenia': df['osteosarcopenia'].sum(),
    'Osteopenia': df['tiene_osteopenia'].sum(),
    'Osteoporosis' : df['tiene_osteoporosis'].sum()
}).sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=conteo_condiciones.index, y=conteo_condiciones.values)
plt.title('Clasificación de pacientes según condición clínica.')
plt.ylabel('Número de pacientes')
plt.xlabel('Condición Clínica')
plt.tight_layout()
plt.savefig("../outputs/objetivo2_condiciones_clinicas.png", dpi=300, bbox_inches='tight')

conteo_condiciones
```

Desde el análisis de datos, no es posible indicar tratamientos específicos; sin embargo, sí es viable **identificar patrones de limitación funcional** que pueden orientar esa decisión médica.

**Variables utilizadas:**

## Objetivo 6: Clasificación de pacientes según niveles de funcionalidad (SPPB).
Se busca clasificar a los pacientes según su nivel de rendimiento físico a partir del puntaje obtenido en la prueba _Short Physical Performance Battery (SPPB)_. Para lo cual, se utilizará el puntaje total (`puntaje_sppb`) y se agrupa a los pacientes en 4 niveles:

- 0 a 4: Limitación grave
- 5 a 6: Limitación moderada
- 7 a 9: Limitación leve
- 10 a 12: Limitación mínima

Esta categorización permitirá explorar la distribución funcional de la muestra.

```python
# creación de la nueva variable categórica

"""
Esta función  toma el valor de cada puntaje__sppb y lo clasifica 
en la categoría correspondiente.
"""
def clasificar_sppb(puntaje):
    if puntaje <= 4:
        return 'Grave'
    elif puntaje <= 6:
        return 'Moderada'
    elif puntaje <= 9:
        return 'Leve'
    else:
        return 'Mínima'

df['sppb_categoria'] = df['puntaje_sppb'].apply(clasificar_sppb)
```

```python
orden_sppb = ['Mínima', 'Leve', 'Moderada', 'Grave']
df['sppb_categoria'] = pd.Categorical(df['sppb_categoria'], categories=orden_sppb, ordered=True)

conteo = df['sppb_categoria'].value_counts().sort_index()

plt.figure(figsize=(8,6))
sns.barplot(x=conteo.index, y=conteo.values, hue=conteo.index, palette='Blues_d', legend=False)
plt.title('Distribución de pacientes según niveles de funcionalidad (SPPB).')
plt.xlabel('\nNivel de limitación funcional')
plt.ylabel('Número de pacientes')
plt.tight_layout()
plt.savefig("../outputs/objetivo6_sppb_categorias.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Cruce con osteosarcopenia
cruce = pd.crosstab(df['sppb_categoria'], df['osteosarcopenia'])

fig, ax = plt.subplots(figsize=(8,5))
cruce.plot(kind='bar', stacked=True, ax=ax)
plt.title('SPPB y presencia de osteosarcopenia.')
plt.xlabel('Categoría funcional (SPPB)')
plt.ylabel('Número de pacientes')
plt.xticks(rotation=35)
plt.legend(title='Osteosarcopenia', labels=['No','Si'])
plt.tight_layout()
plt.savefig("../outputs/objetivo6_sppb_vs_osteosarcopenia.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Objetivo 13: Observar si se presentan medicamentos que predispongan a esta patología.

**Criterio del análisis**  
Se analizaron las varibales `medicamento_1` y `medicamento_2`, que incluyen categorías estandarizadas de uso de medicamentos. Se agruparon por tipo y se creó un indicador binario para detercar si un paciente consume antidiabéticos, antihipertensivos o hipnóticos/ansiolíticos. Luego, se comparó la frecuencia de uso de estos medicamentos entre pacientes con y sin osteosarcopenia, con el objetivo de observar su distribución sin asumir una relación causal.

```python
# unificar medicamentos
df['medicamentos_combinados'] = df[['medicamento_1', 'medicamento_2']].values.tolist()
```

```python
# creación de variables binarias
df['usa_antidiabeticos'] = df['medicamentos_combinados'].apply(
    lambda x: 'antidiabeticos' in x if isinstance(x, list) else False
)
df['usa_antihipertensivos'] = df['medicamentos_combinados'].apply(
    lambda x: 'antihipertensivos' in x if isinstance(x, list) else False
)

df['usa_ansioliticos'] = df['medicamentos_combinados'].apply(
    lambda x: 'hipnoticos o ansioliticos' in x if isinstance(x, list) else False
)
```

```python
#  antidiabéticos por osteosarcopenia
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='osteosarcopenia', hue='usa_antidiabeticos', palette='Blues')
plt.title('Consumo de antidiabéticos según osteosarcopenia.')
plt.xlabel('Osteosarcopenia')
plt.ylabel('Número de pacientes')
plt.legend(title="Uso de antidiabéticos", labels=['No', 'Sí'])
plt.xticks([0,1], ['No', 'Sí'])
plt.tight_layout()
plt.savefig("../outputs/objetivo13_antidiabeticos_vs_osteosarcopenia.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# antihipertensivos y osteosarcopenia
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='osteosarcopenia', hue='usa_antihipertensivos', palette='Blues')
plt.title('Consumo de antihipertensivos según osteosarcopenia.')
plt.xlabel('Osteosarcopenia')
plt.ylabel('Número de pacientes')
plt.legend(title='Uso de antihipertensivos', labels=['No', 'Sí'])
plt.xticks([0,1], ['No', 'Sí'])
plt.tight_layout()
plt.savefig("../outputs/objetivo13_antihipertensivos_vs_osteosarcopenia.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ansiolíticos y osteosarcopenia
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='osteosarcopenia', hue='usa_ansioliticos', palette='Blues')
plt.title('Consumo de ansiolíticos o hipnóticos según osteosarcopenia.')
plt.xlabel('Osteosarcopenia')
plt.ylabel('Número de pacientes')
plt.legend(title='Uso de antihipertensivos', labels=['No', 'Sí'])
plt.xticks([0,1], ['No', 'Sí'])
plt.tight_layout()
plt.savefig("../outputs/objetivo13_ansioliticos_vs_osteosarcopenia.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Objetivo 14: Estratificar el riesgo de fractura con el FRAX en los pacientes con osteosarcopenia.

**Variables utilizadas:**  
- `osteosarcopenia`
- `probabilidad_de_fractura_por_fragilidad`
- `probabilidad_de_fractura_de_cadera`
- `estratificacion_de_riesgo`: baja, moderada, alta, muy alta.

**Criterio de análisis:**
Se utilizan las variables derivadas del modelo FRAX disponibles en el conjunto de datos para analizar si los pacientes con osteosarcopenia presentar mayor riesgo de fractura. Se comparan tanto las probabilidades continuas estimadas por FRAX como su clasificación ordinal. Este análisis permite observar si el grupo con osteosarcopenia está sopbre-representado en los niveles más altos de riesgo de fractura.

```python
# categorizar estatificacion_de_riesgo
orden_frax = ['baja', 'moderada', 'alta', 'muy alta']
df['estratificacion_de_riesgo'] = pd.Categorical(df['estratificacion_de_riesgo'], categories=orden_frax, ordered=True)
```

```python
# Comparar probabilidades FRAX por osteosarcopenia
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='osteosarcopenia', y='probabilidad_de_fractura_por_fragilidad', hue='osteosarcopenia', palette='Blues', legend=False)
plt.title('FRAX: Probabilidad de fractura por fragilidad según osteosarcopenia.')
plt.xlabel('Osteosarcopenia')
plt.ylabel('Probabilidad (%)')
plt.xticks([0,1], ['No', 'Sí'])
plt.tight_layout()
plt.savefig("../outputs/objetivo14_frax_fragilidad_vs_osteosarcopenia.png", dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='osteosarcopenia', y='probabilidad_de_fractura_de_cadera', hue='osteosarcopenia', palette='Blues', legend=False)
plt.title('FRAX: Probabilidad de fractura de cadera según osteosarcopenia.')
plt.xlabel('Osteosarcopenia')
plt.ylabel('Probabilidad (%)')
plt.xticks([0,1], ['No', 'Sí'])
plt.tight_layout()
plt.savefig("../outputs/objetivo14_frax_cadera_vs_osteosarcopenia.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# comparar clasificación ordinal de riesgo por osteosarcopenia
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='estratificacion_de_riesgo', hue='osteosarcopenia', palette='Blues')
plt.title('Clasificación de riesgo FRAX según osteosarcopenia.')
plt.xlabel('Estratificación de riesgo')
plt.ylabel('Número de pacientes')
plt.legend(title='Osteosarcopenia', labels=['No', 'Sí'])
plt.tight_layout()
plt.savefig("../outputs/objetivo14_frax_clasificacion_vs_osteosarcopenia.png", dpi=300, bbox_inches='tight')
plt.show()
```
