# Modelo Predictivo Exploratorio.
Modelos experimentales, para ver tendencias.

```python
# Librerías
# tratamiento de datos
import pandas as pd
import numpy as np

# procesamiento
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# modelado
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score

# utilidades
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga
df = pd.read_csv("../data/cleaned/df_eda.csv")
df.head()
```

```python
# variable objetivo y exclusiones

# eliminación de variable altamente correlacionada con variable objetivo
# para evitar data leakage
df = df.drop(columns=['tiene_sarcopenia'])

# definicion "x" y "y"
X = df.drop(columns=['osteosarcopenia'])
y = df['osteosarcopenia'].astype(int)
```

```python
# Clasificación de columnas por tipo de procesamiento

# variable ordinales
categorias_ordenadas = {
    'clas_imc': ['adecuado', 'sobrepeso', 'obesidad 1', 'obesidad 2', 'obesidad 3'],
    'grasa_resultado': ['normal', 'bueno', 'alta', 'peligrosamente alta'],
    'imme_resultado': ['normal', 'sarcopenia grado 1', 'sarcopenia grado 2'],
    'estratificacion_de_riesgo': ['baja', 'moderada', 'alta', 'muy alta'],
    'clasificacion_de_estado_fisico' : ['bajo', 'intermedio', 'alto'],
    'fuerza': ['ninguna', 'poca', 'mucha'],
    'caminata': ['ninguna', 'poca', 'mucha'],
    'levantarse': ['ninguna', 'poca', 'mucha'],
    'subir_escaleras': ['ninguna', 'poca', 'mucha'],
    'caidas': ['ninguna', '1 - 3 caidas', '4 o mas caidas'],
    'tiempo_de_ejercicio': ['no', '30 - 60 mins', '61 - 120 mins', '121 - 180 mins','mas de 180 mins'],
    'nivel_de_sarcopenia': ['sin sarcopenia', 'sarcopenia leve', 'sarcopenia moderada', 'sarcopenia severa']
}

# variables booleanas
bool_cols = [col for col in X.columns if X[col].dtype == 'bool']

# variables numéricas
num_cols = X.select_dtypes(include=['float64', 'int64']).columns.tolist()

# variables categóricas ordinales
ordinal_cols = [col for col in categorias_ordenadas if col in X.columns]

# variables categóricas nominales (no ordenadas)
cat_cols = list(set(X.select_dtypes(include=['object']).columns) - set(ordinal_cols))

# validación
print("Variables numéricas:", num_cols)
print("Variables ordinales:", ordinal_cols)
print("Variables categóricas nominales:", cat_cols)
print("Variables boobleanas:", bool_cols)
```

```python
# construcción de ColumnTransformer de preprocesamiento

# pipeline para variables numéricas
num_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

# pipeline para variables categóricas
ordinal_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ordinal', OrdinalEncoder(categories=[categorias_ordenadas[col] for col in ordinal_cols]))
])

# pipeline para variables categóricas nominales
nominal_pipeline = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('num', num_pipeline, num_cols),
    ('ord', ordinal_pipeline, ordinal_cols),
    ('cat', nominal_pipeline, cat_cols)
], remainder='passthrough') # pasa bool tal cual
```

```python
# pipeline con regresión logística

model = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000, random_state=42))
])
```

```python
# Validación cruzada

# config validación cruzada con 5 particiones
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# evaluacion de desempeño
scores = cross_val_score(model, X, y, cv=cv, scoring='accuracy')

print("Puntajes de validación cruzada (Accuracy):", scores)
print("Promedio de accuracy:", scores.mean())
```

```python
# entrenar modelo final sobre todo el data set
model.fit(X, y)
```

```python
# Evaluar con clasificación y matriz de confusión (re-entrenado sobre todo el dataset)
# exploración interna

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# predicciones sobre los mismos datos (sólo expliración)
y_pred = model.predict(X)

# Matriz de confusión
cm = confusion_matrix(y, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.named_steps['classifier'].classes_)
disp.plot(cmap='Blues')
plt.title('Matriz de confusión sobre el conjunto completo')
plt.show()

# Reporte de clasificación
print(classification_report(y, y_pred))
```

```python
# Interpretar los coeficientes

# obtener nombres de columnas transformadas
# 1: extraer el preprocesador.
preprocessor = model.named_steps['preprocessing']

# 2: obtener nombres de variables codificadas
# numéricas
num_features = num_cols

# ordinales
ord_features = ordinal_cols

# nominales
cat_encoder = preprocessor.named_transformers_['cat'].named_steps['onehot']
cat_feature_names = cat_encoder.get_feature_names_out(cat_cols)

# booleanas
bool_features = bool_cols

# unir todo
feature_names = np.concatenate([num_features, ord_features, cat_feature_names, bool_features])

# 3: obtener coeficientes del modelo
coefs = model.named_steps['classifier'].coef_[0]

# Crear DataFrame ordenado
coef_df = pd.DataFrame({
    'feature': feature_names,
    'coeficiente': coefs,
    'abs_val': np.abs(coefs)
}).sort_values(by='abs_val', ascending=False)

# visualización
plt.figure(figsize=(10,6))
sns.barplot(data=coef_df.head(15), x='abs_val', y='feature', hue='abs_val', palette='Blues_d', legend=False)
plt.title('Top 15 Variables más influyentes en la predicción de osteosarcopenia')
plt.xlabel('Importancia (|coef|)')
plt.ylabel('Variable')
plt.tight_layout()
plt.savefig('../variables_rl.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Árbol de decisión

from sklearn.tree import DecisionTreeClassifier

# nuevo pipeline
tree_model = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('classifier', DecisionTreeClassifier(max_depth=4, random_state=42))
])
```

```python
# Validación cruzada
tree_scores = cross_val_score(tree_model, X, y, cv=cv, scoring='accuracy')

print("Puntajes de validación cruzada (Accuracy):", tree_scores)
print("Promedio de accuracy:", tree_scores.mean())
```

```python
# entrenamiento
tree_model.fit(X, y)

# extraer árbol entrenado
tree = tree_model.named_steps['classifier']
```

```python
# Obtener e interpretar importancia de variables

# Obtener los nombres de las variables transformadas
feature_names = np.concatenate([num_features, ord_features, cat_feature_names, bool_features])

# Obtener imporatncias
importances = tree.feature_importances_

# crear df ordenado
import_df = pd.DataFrame({
    'feature': feature_names,
    'importancia': importances
}).sort_values(by='importancia', ascending=False)

# viz
plt.figure(figsize=(10,6))
sns.barplot(data=import_df.head(15), x='importancia', y='feature', hue='importancia', palette='Blues_d', legend=False)
plt.title('Top 15 Variables más importantes en el Árbol de Decisión')
plt.xlabel('Importancia')
plt.ylabel('Variable')
plt.tight_layout()
plt.savefig('../variables_tree.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

# Creamos un nuevo pipeline para Random Forest
rf_model = Pipeline(steps=[
    ('preprocessing', preprocessor),
    ('classifier', RandomForestClassifier(
        n_estimators=100,
        max_depth=4,  # Mantenerlo pequeño por tamaño de muestra
        random_state=42
    ))
])
```

```python
# Evaluación con validación cruzada estratificada
rf_scores = cross_val_score(rf_model, X, y, cv=cv, scoring='accuracy')

print(f"Puntajes de validación cruzada (Accuracy): {rf_scores}")
print(f"Promedio de accuracy: {rf_scores.mean():.3f}")
```

```python
# Entrenar sobre todos los datos para interpretación
rf_model.fit(X, y)

# Extraer el modelo Random Forest entrenado
rf = rf_model.named_steps['classifier']
```

```python
# Reutilizamos los nombres de columnas transformadas
rf_importances = rf.feature_importances_

# DataFrame de importancia
rf_import_df = pd.DataFrame({
    'feature': feature_names,
    'importancia': rf_importances
}).sort_values(by='importancia', ascending=False)

# Visualización
plt.figure(figsize=(10, 6))
sns.barplot(data=rf_import_df.head(15), x='importancia', y='feature', hue='importancia', palette='Blues_d', legend=False)
plt.title('Top 15 Variables más importantes según Random Forest')
plt.xlabel('Importancia')
plt.ylabel('Variable')
plt.tight_layout()
plt.savefig("../variables_rf.png", dpi=300, bbox_inches='tight')
plt.show()
```
